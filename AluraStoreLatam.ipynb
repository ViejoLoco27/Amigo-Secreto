#1. Análisis de facturación



````python
t1_ing_totales = sum(tienda['Precio'])
t2_ing_totales = sum(tienda2['Precio'])
t3_ing_totales = sum(tienda3['Precio'])
t4_ing_totales = sum(tienda4['Precio'])

print(f'\nLas ventas totales de la tienda 1 = {t1_ing_totales }')
print(f'\nLas ventas totales de la tienda 2 = {t2_ing_totales}')
print(f'\nLas ventas totales de la tienda 3 = {t3_ing_totales}')
print(f'\nLas ventas totales de la tienda 4 = {t4_ing_totales}')
````

#1.1 Gráfico de barras del análisis de facturación

````python
import matplotlib.pyplot as plt

# Ingresos totales por tienda
t1_ing_totales = sum(tienda['Precio']) / 1000000
t2_ing_totales = sum(tienda2['Precio']) / 1000000
t3_ing_totales = sum(tienda3['Precio']) / 1000000
t4_ing_totales = sum(tienda4['Precio']) / 1000000

# Definir etiquetas y valores
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [t1_ing_totales, t2_ing_totales, t3_ing_totales, t4_ing_totales]

# Crear la gráfica de barras
plt.bar(tiendas, ingresos, color='royalblue')

# Etiquetas y título
plt.xlabel('Tiendas')
plt.ylabel('Ingresos totales')
plt.title(f'Ingresos totales por Tienda\n(millones de pesos)')

# Mostrar valores en las barras
for i, valor in enumerate(ingresos):
    plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Mostrar la gráfica
plt.show()
````

#5. Costo de envio promedio

````python
mean_envio_t1 = sum(tienda['Costo de envío']) / len(tienda['Costo de envío'])
mean_envio_t2 = sum(tienda2['Costo de envío']) / len(tienda2['Costo de envío'])
mean_envio_t3 = sum(tienda3['Costo de envío']) / len(tienda3['Costo de envío'])
mean_envio_t4 = sum(tienda4['Costo de envío']) / len(tienda4['Costo de envío'])

print(f'\nEl consto de envio promedio de la tienda 1 = {round(mean_envio_t1, 2)}')
print(f'\nEl consto de envio promedio de la tienda 2 = {round(mean_envio_t2, 2)}')
print(f'\nEl consto de envio promedio de la tienda 3 = {round(mean_envio_t3, 2)}')
print(f'\nEl consto de envio promedio de la tienda 4 = {round(mean_envio_t4, 2)}')
````

````python
#Operación para obtener el ingreso promedio

t1_ing_promedios = round(sum(tienda['Precio']) / len(tienda),2)
t2_ing_promedios = round(sum(tienda2['Precio']) / len(tienda2),2)
t3_ing_promedios = round(sum(tienda3['Precio']) / len(tienda3),2)
t4_ing_promedios = round(sum(tienda4['Precio']) / len(tienda4),2)

print(f'\nEl ingreso promedio de la tienda 1 = {t1_ing_promedios}')
print(f'\nEl ingreso promedio de la tienda 2 = {t2_ing_promedios}')
print(f'\nEl ingreso promedio de la tienda 3 = {t3_ing_promedios}')
print(f'\nEl ingreso promedio de la tienda 4 = {t4_ing_promedios}')
````

````python
#Operación para obtener la ganacia promedio
ganancia_media1 = round(t1_ing_promedios-mean_envio_t1,2)
ganancia_media2 = round(t2_ing_promedios-mean_envio_t2,2)
ganancia_media3 = round(t3_ing_promedios-mean_envio_t3,2)
ganancia_media4 = round(t4_ing_promedios-mean_envio_t4,2)

print(ganancia_media1)
print(ganancia_media2)
print(ganancia_media3)
print(ganancia_media4)
````
